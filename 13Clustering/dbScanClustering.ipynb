```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
X,Y=datasets.make_moons(n_samples=1000,shuffle=True,noise=.075,random_state=42)
X,Y
```

```python
X=pd.DataFrame(X,columns=['first feature','second feature'])
X.head()
```

```python
plt.scatter(X['first feature'],X['second feature'],c=Y)
```

```python
# Standardize the values
from sklearn.preprocessing import StandardScaler
for column in X.columns:
	scaler=StandardScaler()
	X[column]=scaler.fit_transform(X[[column]])

X.head()
```

```python
from sklearn.cluster import DBSCAN
dbScan=DBSCAN(eps=.25)
dbScan.fit(X)
```

```python
dbScan.labels_
```

```python
plt.scatter(X['first feature'],X['second feature'],c=dbScan.labels_)
```
