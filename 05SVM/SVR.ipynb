```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as ply
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.svm import SVC,SVR
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

- use LabelEncoding for sex,smoker,time
- use One Hot Encoding for 'day'


```python
X=df.iloc[:,1:]
Y=df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.3,random_state=15)
```

```python
label_encoder1=LabelEncoder()
label_encoder2=LabelEncoder()
label_encoder3=LabelEncoder()

X_train['sex']=label_encoder1.fit_transform(X_train['sex'])
X_train['smoker']=label_encoder2.fit_transform(X_train['smoker'])
X_train['time']=label_encoder3.fit_transform(X_train['time'])
```

```python
X_train
```

```python
OHE_encoder=OneHotEncoder()
OHE_encoder.fit_transform(X_train[['day']])
```

```python
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[('OHE_encoder',OneHotEncoder(drop='first'),[3])],remainder='passthrough')
X_train=ct.fit_transform(X_train)
```

```python
X_test
```

```python
X_test['sex']=label_encoder1.transform(X_test['sex'])
X_test['smoker']=label_encoder2.transform(X_test['smoker'])
X_test['time']=label_encoder3.transform(X_test['time'])

X_test=ct.transform(X_test)
```

```python
X_test
```

```python
svr=SVR()
svr.fit(X_train,Y_train)
```

```python
Y_pred=svr.predict(X_test)
```

```python

print(r2_score(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
svr=SVR()
params={
	'C':[.001,.01,1,10,100],
	'kernel':['rbf'],
	'gamma':[.1,6.7,4.99]
}

grid=GridSearchCV(estimator=svr,param_grid=params,cv=5)
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
Y_pred=grid.predict(X_test)
```

```python
print(r2_score(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```
