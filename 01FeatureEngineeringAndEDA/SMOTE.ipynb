### SMOTE (Synthetic Minority Oversampling Technique)
SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
import sklearn.datasets as ds
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,Y=ds.make_classification(n_samples=1000,n_features=2,n_redundant=0,random_state=123,weights=[.9])
```

```python
input_sample=pd.DataFrame(X,columns=['f1','f2'])
output_sample=pd.DataFrame(Y,columns=['val'])
```

```python
df=pd.concat([input_sample,output_sample],axis=1)
```

```python
df
```

```python
df['val'].value_counts()
```

```python
plt.scatter(df['f1'],df['f2'],c=df['val'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
X,Y=oversample.fit_resample(df[['f1','f2']],df["val"])
```

```python
df=pd.concat([X,Y],axis=1)
```

```python
df
```

```python
plt.scatter(df['f1'],df['f2'],c=df['val'])
```
